```python
import pandas as pd
df = pd.read_csv("AmesHousing.csv")
df
```

```python
df.columns
```

```python
df.loc[:, ['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built', 'SalePrice']]
```

```python
missing = df.loc[:,['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built', 'SalePrice']]
missing.isnull().sum()
```

```python
df['Year Built'] = df['Year Built'].astype(int)
df['Year Built'].dtypes
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built']] = scaler.fit_transform(df[['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built']])
df[['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built']]
```

```python
df.dtypes
```

```python
from sklearn.model_selection import train_test_split
x = df[['Gr Liv Area', 'Bedroom AbvGr', 'Full Bath', 'Year Built']]
y = df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE :{mse}, R2 Score :{r2}")
```

```python
coef = pd.DataFrame(model.coef_, x.columns, columns=['Coefficients'])
print(coef)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.scatter(y_test, y_pred)
ax.set(
    xlabel = 'Actual Price',
    ylabel = 'Predicted Price',
    title = 'Comparison of Actual Price with Predicted Price'
)
plt.show()
```
